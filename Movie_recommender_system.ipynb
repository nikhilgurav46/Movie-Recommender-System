```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
print("numpy", np.__version__)
print("pandas", pd.__version__)
```

```python
movies = movies.merge(credits, on = 'title') # Merge the 2 datasets into one
movies.head(3)
```

```python
# Selecting only those features that are important
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies.head(3)
```

```python
movies.info()
```

```python
movies.isnull().sum() # Checking for null values
```

```python
movies.dropna(inplace = True) # There are only 3 null values in dataset so we are Dropping the Null values
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum() # Checking for Duplicated values
```

```python
import ast

# Function to get the Genres and Keywords for that perticular movie
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(3)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head(3)
```

```python
# Function to get only top 3 cast from that movie
def convert2(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L    
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
movies.head(3)
```

```python
# Creating function that Fetches Director's name
def fetch_dir(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_dir) 
```

```python
movies['crew'].head(3)
```

```python
# Converting the List format in overview column into list type
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies['overview'].head(3)
```

```python
# Eliminating the spaces in between that are in genres, keywords, cast and crew column
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies.head(3)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
new_df.head(3)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower()) # Converting all the strings to lower case
new_df.head(3)
```

Using Text Vectorization Technique to get movie that is Related to that of the users intrest.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english') # we will eliminate the stop words like: 'a', 'or', 'to', etc.
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
!pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x: x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins') # Testing
```
